```python
import pickle as pkl
```

```python
Credit = pkl.load(open('Credit.pkl','rb'))
Elite = pkl.load(open('Elite.pkl','rb'))
Supp = pkl.load(open('Supp.pkl','rb'))
```

```python
for i in range(len(Credit)):
    Credit[i].append("Credit")
for i in range(len(Elite)):
    Elite[i].append("Elite")
for i in range(len(Supp)):
    Supp[i].append("Supp")
#     Supp[i][0]*=1.3
#     Elite[i][0]*=1.3
```

```python
lst = []
d={}
d[0]="Supp"
d[1]="Elite"
d[2]="Credit"
for i in range(len(Credit)):
    lst.append(max([Credit[i],Elite[i],Supp[i]]))
```

```python
lst.sort(reverse=True)
len(lst)
```

```python
f = open('42_IITMadras_71.csv',"w")
for i in range(1000):
    f.write(str(lst[i][1])+","+lst[i][2]+"\n")
f.close()
```
