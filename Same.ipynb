```python
import csv
Y=[]
with open('42_IITMadras_30.csv') as f1,open('42_IITMadras_31.csv') as f2:
    r1=csv.reader(f1)
    r2=csv.reader(f2)
    st=set((row[0],row[1]) for row in r1)
    i=0
    for row in r2:
        if i>500: break
        if tuple(row) in st:
            Y.append(row)
        i+=1
```

```python
import csv
Z=[]
with open('42_IITMadras_30.csv') as f1,open('42_IITMadras_31.csv') as f2:
    r1=csv.reader(f1)
    r2=csv.reader(f2)
    st=set(row[0] for row in r1)
    i=0
    for row in r2:
        if i>500: break
        if row[0] in st:
            Z.append(row)
        i+=1
```

```python
Z_Y = []
for e in Z:
    if e not in Y: Z_Y.append(e)
```
